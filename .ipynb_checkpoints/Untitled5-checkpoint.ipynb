```python
#設定
import jieba
import jieba.analyse
import pandas as pd
import numpy as np
from collections import Counter
orgdata = pd.read_csv('COA_OpenData.csv')
```

```python
orgdata['seg'] = " "
orgdata.head()
```

```python
jieba.analyse.set_stop_words('stopwords.txt')
index = 0
for documents in orgdata['FoodFeature']:
    seg_list = jieba.analyse.extract_tags(documents, 150)
    #seg_list = jieba.lcut(documents)
    orgdata['seg'].iloc[index] = seg_list
    index += 1
```

```python
orgdata['seg']
```

```python
#把city整理一下
Counter(orgdata['City']).keys()
```

```python
# 建立一個 DataFrame 兩欄位分別放入orgdata['City']的keys跟values
show_key = pd.DataFrame(Counter(orgdata['City']).keys())
show_values = pd.DataFrame(Counter(orgdata['City']).values())
df = [show_key, show_values]

#  合併資料並加上欄位名稱
result = pd.concat(df, axis=1)
result.columns = ['City', 'times']
result
```

```python
#把資料轉成圖表
import plotly.express as px
import pandas as pd
px.bar(result, x='City', y='times', title='the number of taiwan each city attractions ')
```

```python
#因為台南最多所以特別看台南
show_sub = orgdata['City']=='臺南市'
subdata = orgdata[show_sub]
subdata
```

```python
subdata_seg_list = subdata['seg'].tolist()
subdata_seg_list
```

```python
#把台南的資料轉成文字雲
flat = [x for sublist in subdata_seg_list for x in sublist]
diction = Counter(flat)

from PIL import Image
import matplotlib.pyplot as plt
from wordcloud import WordCloud

mask = np.array(Image.open("map.png"))
font = "/System/Library/Fonts/STHeiti Medium.ttc"

wordcloud = WordCloud(background_color = "white", font_path=font, mask = mask)
wordcloud.generate_from_frequencies(diction)

plt.figure(figsize=(20,20))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
